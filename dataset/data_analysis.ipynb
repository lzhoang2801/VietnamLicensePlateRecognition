```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import random
```

# A. Vietnam License Plate Segment Datasets

About dataset: The license plate zoning dataset in Vietnam is divided into 2 classes, 1-line number plates (LpD) and 2-line number plates (LpV). Total images we have collected about 5,000 images, 1-line license plate sample has 3,510 samples and 2-line license plate has 1,625

Source: https://www.kaggle.com/datasets/duydieunguyen/licenseplates

### Sample Visualization

```python
dataset_dir = "licenseplates/images/train"
num_samples = 36

image_files = os.listdir(dataset_dir)

random.shuffle(image_files)

plt.figure(figsize=(20, 20))
for i, image_file in enumerate(image_files):
    image_path = os.path.join(dataset_dir, image_file)
    image = cv2.imread(image_path)

    plt.subplot(6, 6, i + 1)
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')

    num_samples -= 1
    if num_samples == 0:
        break
plt.show()
```

# B. Digit and English Character Dataset

About dataset: It contains 442,450 images, each sized 28x28 pixels, and a label column. There are 36 possible labels, corresponding to the digits 0 to 9 and the letters A to Z (where A is 10 and Z is 35).

Source: https://www.kaggle.com/datasets/swarupshah/digit-character

```python
df = pd.read_csv("digit_char_dataset.csv")
```

### Character Distribution

```python
def class_to_label(x):
    if 0 <= x <= 9:
        return str(x)
    elif 10 <= x <= 35:
        return chr(ord('A') + x - 10)
    else:
        return str(x)

labels = df['class'].apply(class_to_label)

plt.figure(figsize=(20, 6))
labels.value_counts().sort_index(key=lambda x: [int(i) if i.isdigit() else ord(i) for i in x]).plot(kind='bar')
plt.xlabel('Character')
plt.ylabel('Number of Images')
plt.title('Distribution of Images per Character')
plt.show()
```

### Sample Character Visualization

```python
num_samples = 6
plt.figure(figsize=(12, 12))
chars = df['class'].unique()[20:26]
for c_idx, char in enumerate(chars):
    images_rows = df[df['class'] == char].iloc[:num_samples]
    images = images_rows.drop('class', axis=1).to_numpy()
    images = images.reshape(images.shape[0], 28, 28).astype('uint8')

    for i, image in enumerate(images):
        plt.subplot(len(chars), num_samples, c_idx * num_samples + i + 1)
        plt.imshow(image, cmap='gray')
        plt.axis('off')
        if i == 0:
            plt.title(class_to_label(char))
plt.tight_layout()
plt.show()
```
