```python
import cv2
import matplotlib.pyplot as plt
import os
import sys

sys.path.append(os.path.abspath('..'))

from utils.yolo_to_haar import get_haar_annotations
```

```python
image_path = "../dataset/licenseplates/images/train/Dieu_0171.png"
label_path = "../dataset/licenseplates/labels/train/Dieu_0171.txt"

image = cv2.imread(image_path)
label = open(label_path, 'r').read()
```

```python
haar_data = get_haar_annotations(image_path, label_path)

for data in haar_data:
    cv2.rectangle(image, (data[2], data[3]), (data[2] + data[4], data[3] + data[5]), (0, 0, 255), 2)

plt.imshow(image)
plt.show()
```
